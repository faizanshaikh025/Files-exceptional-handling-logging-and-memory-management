```python
#5.How would you catch and handle division by zero error in Python?
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return float('nan')  # Return 'Not a Number' on error
    except TypeError:
        return "Error: Invalid input types."

print(safe_divide(10, 0))    # Output: nan
print(safe_divide(10, 2))    # Output: 5.0
print(safe_divide(10, "a")) # Output: Error: Invalid input types.
```

```python
#8. Write a program to handle a file opening error using exception handling.
import os

def read_file_safely(filename):
    """
    Attempts to open and read a file with comprehensive error handling.
    Returns the file content if successful, None otherwise.
    """
    try:
        # Check if file exists first (optional - just for demonstration)
        if not os.path.exists(filename):
            raise FileNotFoundError(f"The file '{filename}' does not exist")

        # Attempt to open and read the file
        with open(filename, 'r') as file:
            content = file.read()
            print(f"Successfully read file: {filename}")
            return content

    except FileNotFoundError as e:
        print(f"Error: {e}")
        print("Please check the file name and try again.")
    except PermissionError:
        print(f"Error: Permission denied when trying to read '{filename}'")
        print("Please check your file permissions.")
    except IsADirectoryError:
        print(f"Error: '{filename}' is a directory, not a file")
    except UnicodeDecodeError:
        print(f"Error: Could not decode the contents of '{filename}'")
        print("The file may be a binary file or use a different encoding.")
    except IOError as e:
        print(f"Input/Output error occurred: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    finally:
        print(f"File handling attempt completed for: {filename}")

    return None

# Example usage
if __name__ == "__main__":
    file_path = "example.txt"  # Change this to test different scenarios

    # Test cases you can try:
    # 1. A file that exists - will read successfully
    # 2. A non-existent file - FileNotFoundError
    # 3. A directory - IsADirectoryError
    # 4. A file without read permissions - PermissionError

    print("\nAttempting to read file...")
    file_content = read_file_safely(file_path)

    if file_content:
        print("\nFile content preview:")
        print(file_content[:200] + "..." if len(file_content) > 200 else file_content)
```

```python
#11.Write a Python program that uses a try-except block to handle an error when attempting to access a dictionary key that doesn't exist.
def get_dict_value(dictionary, key):
    """
    Safely retrieves a value from a dictionary with error handling.
    Returns the value if key exists, None otherwise.
    """
    try:
        value = dictionary[key]
        print(f"Found key '{key}': {value}")
        return value
    except KeyError:
        print(f"Warning: Key '{key}' does not exist in the dictionary")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

# Example usage
if __name__ == "__main__":
    # Sample dictionary
    user_data = {
        "name": "Alice",
        "age": 30,
        "email": "alice@example.com"
    }

    # Test cases
    print("=== Safe Dictionary Access ===")
    name = get_dict_value(user_data, "name")       # Existing key
    address = get_dict_value(user_data, "address")  # Non-existent key
    age = get_dict_value(user_data, "age")         # Existing key

    # Alternative approach using dict.get()
    print("\n=== Using dict.get() ===")
    phone = user_data.get("phone", "N/A")  # Returns "N/A" if key doesn't exist
    print(f"Phone number: {phone}")
```

```python
#14.Write a program that uses the logging module to log both informational and error messages.
import logging
import os

def configure_logging():
    """Set up logging configuration with both file and console output"""
    # Create logs directory if it doesn't exist
    os.makedirs('logs', exist_ok=True)

    # Configure the root logger
    logging.basicConfig(
        level=logging.INFO,
        format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
        handlers=[
            logging.FileHandler('logs/app.log'),  # Log to file
            logging.StreamHandler()              # Log to console
        ]
    )

def process_data(data):
    """Example function that processes data and logs events"""
    logger = logging.getLogger('data_processor')

    try:
        logger.info(f"Starting to process data: {data}")

        # Simulate processing
        if not isinstance(data, (int, float)):
            logger.warning("Non-numeric data received, attempting conversion")
            data = float(data)

        result = 100 / data
        logger.info(f"Processing successful. Result: {result}")
        return result

    except ValueError as e:
        logger.error(f"Failed to convert data: {e}", exc_info=True)
        return None
    except ZeroDivisionError:
        logger.critical("Division by zero attempted!")
        return None
    except Exception as e:
        logger.exception("Unexpected error occurred during processing")
        return None

def main():
    """Main program execution"""
    configure_logging()
    logger = logging.getLogger('main')

    logger.info("Program started")

    # Test cases
    test_data = [10, "5", 0, "abc"]

    for data in test_data:
        logger.debug(f"Processing test case: {data}")  # Won't show with INFO level
        result = process_data(data)
        if result is not None:
            logger.info(f"Got result: {result}")
        else:
            logger.warning("No result returned")

    logger.info("Program completed")

if __name__ == "__main__":
    main()
```

```python
#18.How would you implement a basic logging setup that logs to a file with rotation after 1MB?
import logging
from logging.handlers import RotatingFileHandler

def setup_logging(log_file='app.log', max_size=1, backup_count=5):
    """
    Set up logging with file rotation

    Args:
        log_file (str): Path to log file
        max_size (int): Max file size in MB before rotation
        backup_count (int): Number of backup logs to keep
    """
    # Create logger
    logger = logging.getLogger()
    logger.setLevel(logging.INFO)  # Set minimum log level

    # Create formatter
    formatter = logging.Formatter(
        '%(asctime)s - %(name)s - %(levelname)s - %(message)s'
    )

    # Create rotating file handler
    file_handler = RotatingFileHandler(
        filename=log_file,
        maxBytes=max_size * 1024 * 1024,  # Convert MB to bytes
        backupCount=backup_count,
        encoding='utf-8'
    )
    file_handler.setFormatter(formatter)

    # Add handler to logger
    logger.addHandler(file_handler)

    # Also log to console (optional)
    console_handler = logging.StreamHandler()
    console_handler.setFormatter(formatter)
    logger.addHandler(console_handler)

    return logger

# Example usage
if __name__ == "__main__":
    logger = setup_logging('my_app.log')

    logger.info("Application started")
    logger.warning("This is a warning message")
    logger.error("This is an error message")

    # Generate enough logs to trigger rotation
    for i in range(1000):
        logger.info(f"Log entry {i}")
```

```python
#19.Write a program that handles both IndexError and KeyError using a try-except block.
def handle_data_structures(data_list, data_dict, list_index, dict_key):
    """
    Demonstrates handling both IndexError and KeyError

    Args:
        data_list (list): A list to access by index
        data_dict (dict): A dictionary to access by key
        list_index (int): Index to try accessing in list
        dict_key (str): Key to try accessing in dictionary
    """
    try:
        # Try accessing both data structures
        list_value = data_list[list_index]
        dict_value = data_dict[dict_key]

        print(f"List value at index {list_index}: {list_value}")
        print(f"Dictionary value for key '{dict_key}': {dict_value}")

        return list_value, dict_value

    except IndexError:
        print(f"Error: List index {list_index} out of range (list length: {len(data_list)})")
    except KeyError:
        print(f"Error: Dictionary key '{dict_key}' not found (available keys: {list(data_dict.keys())})")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    finally:
        print("Data access attempt completed")

# Example usage
if __name__ == "__main__":
    # Sample data
    my_list = [10, 20, 30]
    my_dict = {'a': 1, 'b': 2, 'c': 3}

    print("=== Test Case 1: Valid Access ===")
    handle_data_structures(my_list, my_dict, 1, 'b')

    print("\n=== Test Case 2: IndexError ===")
    handle_data_structures(my_list, my_dict, 5, 'b')

    print("\n=== Test Case 3: KeyError ===")
    handle_data_structures(my_list, my_dict, 1, 'x')

    print("\n=== Test Case 4: Both Errors ===")
    handle_data_structures(my_list, my_dict, 5, 'x')
```

```python
#21.Write a Python program that reads a file and prints the number of occurrences of a specific word.
def count_word_occurrences(filename, target_word):
    count = 0
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            words = line.strip().split()
            count += sum(1 for word in words if word.lower() == target_word.lower())
    return count

# Example usage
filename = 'example.txt'
word_to_count = 'python'

occurrences = count_word_occurrences(filename, word_to_count)
print(f"The word '{word_to_count}' occurs {occurrences} times in '{filename}'.")
```
